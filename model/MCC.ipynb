```python
import pandas as pd

df = pd.DataFrame(list(zip(list1, teamPosition)),
               columns =['team', 'position'])
df
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

pdlist1 = pd.Series(list1)


mlb = MultiLabelBinarizer()

res = pd.DataFrame(mlb.fit_transform(pdlist1),
                   columns=mlb.classes_,
                   index=pdlist1.index)
```

```python
res
```

```python
res.to_numpy()[0]
```

```python
# Create X & y values
X = res
y = df["position"]
```

```python
# Create training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.1, 
                                                    random_state=42) # set random state for reproducible splits
```

# Training the model

```python
import tensorflow as tf
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a new model (same as model_2)
mcc_model = tf.keras.Sequential([
  tf.keras.layers.Dense(4),
  tf.keras.layers.Dense(300),
  tf.keras.layers.Dense(1000),
  tf.keras.layers.Dense(300),
  tf.keras.layers.Dense(1)
])

# Compile the model
mcc_model.compile(loss=tf.keras.losses.mae,
                        optimizer=tf.keras.optimizers.SGD(),
                        metrics=['mae'])

# Fit the model
mcc_model.fit(X_train, y_train, epochs=90)
```

# Evaluating the model

```python
mcc_model.evaluate(X_test, y_test)
```
